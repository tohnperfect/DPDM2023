### replace missing value with mean of the sub-group

````python
set(unit_df['propertytype_name_en'])
````

#### for Apartment

1. filter only Apartment data

````python
unit_df[unit_df['propertytype_name_en']=='Apartment'].head()
````

2. calculate the average price_min of the group

````python
unit_df[unit_df['propertytype_name_en']=='Apartment']['price_min'].mean()
````

3. Replace the mean to the missing values of the Apartment group

````python
unit_df_apart = unit_df[unit_df['propertytype_name_en']=='Apartment'].fillna(
    {'price_min' : unit_df[unit_df['propertytype_name_en']=='Apartment']['price_min'].mean()})
````

````python
unit_df_apart.isnull().any()
````

#### ทำ Condo

````python
unit_df_condo = unit_df[unit_df['propertytype_name_en']=='Condo'].fillna(
    {'price_min' : unit_df[unit_df['propertytype_name_en']=='Condo']['price_min'].mean()})
````

````python
unit_df_condo.head()
````

````python
unit_df_condo.describe()
````

````python
unit_df_apart.head()
````

## concatenate tables in the Y-axis (add more data points)

````python
fused_df = pd.concat([unit_df_apart,unit_df_condo])
````

### Dict-Row

````python
sale = [{'account':'Jones LLC','Jan':150,'Feb':200,'Mar':140},
        {'account':'Alpha Co','Jan':200,'Feb':210,'Mar':215},
        {'account':'Blue Inc','Jan':50,'Feb':90,'Mar':95}
]

sale_tb1 = pd.DataFrame(sale)
sale_tb1
````

Dict-Column

````python
sales = { 'account':['Jones LLC','Alpha Co','Blue Inc'],
          'Jan'    :[150,200,50],
          'Feb'    :[200,210,90],
          'Mar'    :[140,215,95]
}
sale_tb2 = pd.DataFrame.from_dict(sales)
sale_tb2
````

````python
sale = [['Jones LLC',150,200,140],
        ['Alpha Co',200,210,215],
        ['Blue Inc',50,90,95]]
labels = ['account','Jan','Feb','Mar']

sale_tb3 = pd.DataFrame.from_records(sale, columns=labels)
sale_tb3
````
